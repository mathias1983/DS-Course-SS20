```python
import pandas as pd
import numpy as np

#1) Have a look at the following dataset: dsm-beuth-edl-demodata-orig
# Write a Python / Panda Script which 'cleans' this data set. Justify your actions in the respective notebook or python script you provide as a solution (link, file, kaggle repo, etc.).
# The original dataset does not necessarily have to be created. A proper strategy / good arguments are more important. Value: 5 points.

df = pd.read_csv("data.csv")
#col names which carry user data
data_cols = ['email', 'last_name', 'first_name', 'gender', 'age']

df
```

```python
# Delete empty rows 
#
# All rows without value. 
# Exclude column 'id' because it is just an autoincrement and can be restored later.
df = df.dropna(subset=data_cols, how='all')
df = df.reset_index(drop=True)

df
```

```python
# Delete redundant 'full_name' column.
#
# ... to be more consistent.
# This might be unnecessary here. We could just report inconsistencies, fix them and keep 'full_name' 
# as helper column.
# But here just for the fun: Drop column 'full_name' if data is completly redundant.
# Otherwise display inconsistent rows and do nothing.
#
is_full_name_redundant = df[df.full_name == df.first_name +  ' '  + df.last_name].shape[0] == df.shape[0]
if is_full_name_redundant:
    print("drop redundant column 'full_name'")
    df = df.drop("full_name", axis=1)
else:
    print("Inconsistency with person naming columns listed below")
    df = df[df.full_name != df.first_name + ' ' + df.last_name]
df
```

```python
# Reorder the columns by uniqueness.
#
# This might not be necessary but could be useful.
# Email is a good unique identifier before 'last_name' and 'first_name'. 
#
#
df = df.reindex(columns=['id','email', 'last_name', 'first_name', 'gender', 'age']).sort_values(by=['email'])
df
```

```python
# Show duplicated persons
#
# ... by comparing each entire row (id excluded).
df[df.duplicated(subset=data_cols)]
```

```python
# Delete duplicated persons.
#
# ... by comparing each entire row (id excluded).
# Result: one row deleted
df = df.drop_duplicates(subset=data_cols, keep='last')
df
```

```python
# fix index
# 
df = df.reset_index(drop=True)
df['id'] = range(0, len(df.index))

df
```

```python
# Show null values in column age
#
# ... which we have not
df[df.age.isnull()]
```

```python
# Show non digit values for age column
#
df_ = df[df.age.str.isdigit() == False]
df_
```

```python
# Fix age values and cast column to Int
#
# remove all non digits
df['age'] = df['age'].replace(to_replace=r'[^0-9]', value='', regex=True)
# replace empty values with '0'
# later we can search for value '0' and ask the person to give us this information or something like that
df.loc[df.age.str.isdigit() == False, 'age'] = 0
# cast col to int
df['age'] = df['age'].astype(np.int64)
df
```

```python
# Show persons by age greater than 150
#
# Display possible mistakes
df[df.age > 150]
```
